# DFS

```python
def dfs(graph, start, goal):
    visited = []
    queue = []
    
    visited.append(start)
    queue.append(start)
    
    while queue:
        element = queue.pop(0)
        print(f"{element} ")
        if element == goal:
            break
        
        for n in graph[element]:
            if n not in visited:
                visited.append(n)
                queue.append(n)

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': []
}

dfs(graph, 'A', 'I')
```
