# Q1

```python
import heapq

class DLSGoalAgent:
    def __init__(self, target):
        self.target = target

    def search_graph(self, current_node, graph_data, max_depth, current_path):
        current_path.append(current_node)
        
        if current_node == self.target:
            return current_path
            
        if max_depth <= 0:
            current_path.pop()
            return None
            
        for neighbor in graph_data.get(current_node, []):
            if neighbor not in current_path:
                result = self.search_graph(neighbor, graph_data, max_depth - 1, current_path)
                if result:
                    return result
                    
        current_path.pop()
        return None

class UCSUtilityAgent:
    def __init__(self, target):
        self.target = target

    def find_optimal_path(self, start_node, graph_costs):
        frontier = [(0, start_node, [start_node])]
        explored = set()

        while frontier:
            total_cost, current_node, path = heapq.heappop(frontier)
            
            if current_node == self.target:
                return path, total_cost
                
            if current_node not in explored:
                explored.add(current_node)
                
                for neighbor, step_cost in graph_costs.get(current_node, {}).items():
                    if neighbor not in explored:
                        new_cost = total_cost + step_cost
                        heapq.heappush(frontier, (new_cost, neighbor, path + [neighbor]))
                        
        return None, float('inf')
```

# Q2

```python
class GoalAgent:
    def __init__(self, goal):
        self.goal = goal

    def dls(self, node, graph, limit, path):
        path.append(node)
        
        if node == self.goal:
            return path
            
        if limit == 0:
            path.pop()
            return None
            
        for neighbor in graph.get(node, []):
            if neighbor not in path:
                result = self.dls(neighbor, graph, limit - 1, path)
                if result:
                    return result
                    
        path.pop()
        return None

class UtilityAgent:
    def __init__(self, goal):
        self.goal = goal

    def ucs(self, start, graph):
        queue = [[0, start, [start]]]
        visited = []

        while queue:
            queue.sort(key=lambda x: x[0])
            cost, current, path = queue.pop(0)

            if current == self.goal:
                return path, cost

            if current not in visited:
                visited.append(current)

                for neighbor, weight in graph.get(current, {}).items():
                    if neighbor not in visited:
                        new_cost = cost + weight
                        new_path = path + [neighbor]
                        queue.append([new_cost, neighbor, new_path])
                        
        return None, 0

def get_all_paths(cities):
    if len(cities) <= 1:
        return [cities]
        
    paths = []
    for i in range(len(cities)):
        current = cities[i]
        remaining = cities[:i] + cities[i+1:]
        
        for p in get_all_paths(remaining):
            paths.append([current] + p)
            
    return paths

def solve_tsp(graph, start):
    cities = list(graph.keys())
    cities.remove(start)
    
    all_paths = get_all_paths(cities)
    
    best_path = []
    min_cost = 999999

    for path in all_paths:
        full_path = [start] + path + [start]
        cost = 0
        valid = True
        
        for i in range(len(full_path) - 1):
            c1 = full_path[i]
            c2 = full_path[i+1]
            
            if c2 in graph[c1]:
                cost += graph[c1][c2]
            else:
                valid = False
                break
                
        if valid and cost < min_cost:
            min_cost = cost
            best_path = full_path
            
    return best_path, min_cost

distances = {
    1: {2: 10, 3: 15, 4: 20},
    2: {1: 10, 3: 35, 4: 25},
    3: {1: 15, 2: 35, 4: 30},
    4: {1: 20, 2: 25, 3: 30}
}

def simple_dls(node, target, graph, limit, path):
    path.append(node)
    
    if node == target:
        return path
        
    if limit == 0:
        path.pop()
        return None
        
    for neighbor in graph.get(node, []):
        if neighbor not in path:
            result = simple_dls(neighbor, target, graph, limit - 1, path)
            if result:
                return result
                
    path.pop()
    return None

def iddfs(start, target, graph, max_depth):
    for depth in range(max_depth + 1):
        path = []
        result = simple_dls(start, target, graph, depth, path)
        if result:
            return result
            
    return None
```

# Q3

```python
def depth_limited_dfs(current_node, target_node, structure, limit, visited_path):
    visited_path.append(current_node)
    
    if current_node == target_node:
        return visited_path
        
    if limit <= 0:
        visited_path.pop()
        return None
        
    for adjacent_node in structure.get(current_node, []):
        if adjacent_node not in visited_path:
            found_path = depth_limited_dfs(adjacent_node, target_node, structure, limit - 1, visited_path)
            if found_path:
                return found_path
                
    visited_path.pop()
    return None

def iterative_deepening_search(start_point, target_point, structure, max_allowed_depth):
    for current_depth in range(max_allowed_depth + 1):
        path_tracker = []
        successful_path = depth_limited_dfs(start_point, target_point, structure, current_depth, path_tracker)
        
        if successful_path:
            return successful_path
            
    return None
```
