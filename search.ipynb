```python
#BFS

def BFS(graph, start, goal):
    visited = []
    queue = []
    queue.append(start)
    visited.append(start)

    while queue:
        node = queue.pop(0)
        if node == goal:
            print(f"Goal Node {node} Found")
            print("Path Followed:", end=" ")
            for i in visited:
                print(f"{i}", end=" ")
            return
        for n in graph[node]:
            if n not in visited:
                queue.append(n)
                visited.append(n)

    print(f"Goal Node {goal} Not Found")

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['G'],
    'D': [],
    'E': [],
    'G': ['F'],
    'F': [],
}
BFS(graph, 'A', 'F')
```

```python
# DFS

def DFS(graph, start, goal):
    visited = []
    stack = []
    visited.append(start)
    stack.append(start)

    while stack:
        node = stack.pop()
        print(node, end=" ")
        if node == goal:
            print(f"\nGoal Node {node} Found") 
            return
        for n in reversed(graph[node]):
            if n not in visited:
                stack.append(n)
                visited.append(n)
    print(f"Goal Node {goal} Not Found")

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': []
}

DFS(graph, 'A', 'I')

```

```python
# DLS

def DLS(graph, start, goal, limit):
    visited = []
    stack = []
    stack.append((start, 0))
    visited.append(start)

    while stack:
        node, depth = stack.pop()
        if node == goal:
            print(f"Goal Node {node} Found")
            return True
        if depth < limit:
            for n in graph[node]:
                if n not in visited:
                    visited.append(n)
                    stack.append((n, depth + 1))

    print(f"Goal Node {goal} not found within Limit: {limit}")
    return False


# A simple graph represented as an adjacency list
# Depth 0: A
# Depth 1: B, C
# Depth 2: D, E, F, G
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

# Case 1: Goal is within the limit (Success)
DLS(graph, 'A', 'D', 2)

# Case 2: Goal is OUTSIDE the limit (Failure)
DLS(graph, 'A', 'D', 1)

# Case 3: Goal doesn't exist at all
DLS(graph, 'A', 'Z', 5)
```

```python
# GREEDY BEST-FIRST SEARCH (GBFS)

def greedy_bfs(graph, heuristic, start, goal):
    visited = []
    frontier = [(start, heuristic[start])]

    while frontier:
        # Sort by lowest heuristic f(n) = h(n)
        frontier.sort(key=lambda x: x[1]) 
        node, _ = frontier.pop(0)

        if node == goal:
            print(f"\nGoal Node {node} Found!")
            return True

        if node not in visited:
            print(node, end=" ") # Prints the traversal trace
            visited.append(node)

            for n in graph[node]:
                if n not in visited:
                    frontier.append((n, heuristic[n]))

    print(f"\nNode {goal} Not Found")
    return False


# Graph with edge costs (GBFS ignores the costs, only looks at connections)
graph = {
    'A': {'B': 2, 'C': 1},
    'B': {'D': 4, 'E': 3},
    'C': {'F': 1, 'G': 5},
    'D': {'H': 2},
    'E': {}, 'F': {'I': 6}, 
    'G': {}, 'H': {}, 'I': {}
}

# Heuristic Dictionary (The GPS guess to reach 'I')
heuristic = {'A': 7, 'B': 6, 'C': 5, 'D': 4, 'E': 7, 'F': 3, 'G': 6, 'H': 2, 'I': 0}

# Run the search
print("GBFS Traversal:")
greedy_bfs(graph, heuristic, 'A', 'I')
```

```python
# A* SEARCH

def a_star(graph, h, start, goal):
    visited = []
    queue = []
    
    # NEW: Keep track of the actual cost to reach each node
    g_costs = {start: 0} 

    # Queue stores: (node, f_cost)
    queue.append((start, h[start]))

    while queue:
        # Sort by lowest f(n)
        queue.sort(key=lambda x: x[1])
        node, f_cost = queue.pop(0)

        if node == goal:
            print(f"\nGoal Node {node} Found!")
            return True

        if node not in visited:
            print(node, end=" ")
            visited.append(node)

            # In A*, we NEED the edge weights, so we use .items()
            for neighbour, edge_cost in graph[node].items():
                
                # Calculate exact cost to reach the neighbor
                new_g = g_costs[node] + edge_cost
                
                # If it's a new node OR we found a cheaper path to it
                if neighbour not in g_costs or new_g < g_costs[neighbour]:
                    g_costs[neighbour] = new_g # Save the new cost
                    
                    # f(n) = g(n) + h(n)
                    f_n = new_g + h[neighbour]
                    queue.append((neighbour, f_n))
                    
    print(f"\nNode {goal} Not Found")
    return False


# Graph with actual edge costs
graph = {
    'A': {'B': 4, 'C': 3},
    'B': {'E': 12, 'F': 5},
    'C': {'D': 7, 'E': 10},
    'D': {'E': 2},
    'E': {'G': 5},
    'F': {'G': 16},
    'G': {}
}

# Heuristic Dictionary
heuristic = {'A': 14, 'B': 12, 'C': 11, 'D': 6, 'E': 4, 'F': 11, 'G': 0}

# Run the search
print("A* Search Traversal:")
a_star(graph, heuristic, 'A', 'G')
```
