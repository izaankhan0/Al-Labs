Question 1
A security robot is deployed in a warehouse to prevent theft. The robot acts as the MAX player and the thief acts as the MIN player. The robot chooses Zone A or Zone B, then the thief chooses Option1 or Option2. Use minimax to find the outcome and print all states and moves.

```python
robot_tree = {
    "Zone A": {"Option1": 3, "Option2": 5},
    "Zone B": {"Option1": 2, "Option2": 9},
}
robot_results = {}
print("Task 1: Robot vs Thief")
for robot_move, options in robot_tree.items():
    thief_choice = min(options, key=options.get)
    robot_results[robot_move] = options[thief_choice]
    print(f"Robot chooses {robot_move} -> Thief chooses {thief_choice} with value {options[thief_choice]}")
best_robot_move = max(robot_results, key=robot_results.get)
print("Best robot move:", best_robot_move, "with minimax value", robot_results[best_robot_move])
```

Question 2
A self-driving car approaches an intersection and must decide Stop, Go, or Turn. The car is the MAX player and the aggressive driver is the MIN player. Each action leads to two outcomes. Apply minimax to determine the best action, then apply alpha-beta pruning and identify pruned branches and nodes not evaluated.

```python
car_tree = {
    "Stop": [6, 7],
    "Go": [2, 4],
    "Turn": [8, 3],
}
car_minimax = {action: min(outcomes) for action, outcomes in car_tree.items()}
print("Task 2: Self-driving car")
for action, value in car_minimax.items():
    print(f"{action}: minimax value = {value}")
best_car_action = max(car_minimax, key=car_minimax.get)
print("Best action by minimax:", best_car_action, "with value", car_minimax[best_car_action])
alpha = float("-inf")
pruned_nodes = 0
for action, outcomes in car_tree.items():
    branch_value = float("inf")
    print(f"Entering branch {action} with alpha={alpha}")
    for i, value in enumerate(outcomes):
        print(f"  Evaluating leaf {value}")
        branch_value = min(branch_value, value)
        if branch_value <= alpha:
            pruned_nodes += len(outcomes) - (i + 1)
            print(f"  Prune remaining leaves in {action} after leaf {value}")
            break
    alpha = max(alpha, branch_value)
    print(f"  Branch {action} returns {branch_value}")
print("Nodes not evaluated due to pruning:", pruned_nodes)
```

```python
strategy_tree = {
    "Attack": {"A1": [5, 6], "A2": [7, 4]},
    "Defend": {"D1": [3, 8], "D2": [6, 2]},
    "Gather": {"G1": [1, 9], "G2": [4, 7]},
}
strategy_values = {}
print("Task 3: Turn-based strategy game")
for strategy, responses in strategy_tree.items():
    response_values = {}
    for response, outcomes in responses.items():
        response_values[response] = max(outcomes)
        print(f"{strategy} -> {response}: max({outcomes}) = {response_values[response]}")
    strategy_values[strategy] = min(response_values.values())
    print(f"{strategy}: minimax value = {strategy_values[strategy]}")
best_strategy = max(strategy_values, key=strategy_values.get)
print("Best strategy by minimax:", best_strategy, "with value", strategy_values[best_strategy])
alpha = float("-inf")
pruned_task3 = 0
for strategy, responses in strategy_tree.items():
    current_min = float("inf")
    print(f"Entering strategy {strategy} with alpha={alpha}")
    for response, outcomes in responses.items():
        current_max = float("-inf")
        print(f"  Opponent response {response}")
        for i, value in enumerate(outcomes):
            print(f"    Evaluating leaf {value}")
            current_max = max(current_max, value)
            if current_max >= current_min:
                pruned_task3 += len(outcomes) - (i + 1)
                print(f"    Prune remaining leaves in {response} after leaf {value}")
                break
        print(f"  {response} returns {current_max}")
        current_min = min(current_min, current_max)
        if current_min <= alpha:
            print(f"  Prune remaining responses in {strategy} because current_min={current_min} <= alpha={alpha}")
            break
    print(f"  Strategy {strategy} returns {current_min}")
    alpha = max(alpha, current_min)
print("Alpha-beta pruned leaves:", pruned_task3)
minimax_leaves = sum(len(outcomes) for responses in strategy_tree.values() for outcomes in responses.values())
print("Minimax leaf evaluations:", minimax_leaves)
print("Alpha-beta leaf evaluations:", minimax_leaves - pruned_task3)
```
